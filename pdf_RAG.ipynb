```python
import os
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
import random

# Initialize Chrome options
chrome_options = Options()
chrome_options.add_argument("--start-maximized")
chrome_options.add_argument("--headless")  # Optional: Run in headless mode
chrome_options.add_argument("--disable-gpu")

# Base URL of the website
base_url = 'https://www.doe.go.th/prd/lmia/knowledge/param/site/131/cat/22/sub/0/pull/category/view/cover-view?page='

# Directory to save downloaded PDFs
download_dir = 'DSI444/download_pdfs'
os.makedirs(download_dir, exist_ok=True)

# Set up Selenium WebDriver
service = Service("C:/Users/beamoow/.wdm/drivers/chromedriver/win64/131.0.6778.85/chromedriver.exe")
driver = webdriver.Chrome(service=service, options=chrome_options)

# Iterate through the pages
for page_num in range(6,9):  # Adjust the range as needed
    try:
        url = f"{base_url}{page_num}"
        print(f"Processing page: {url}")
        driver.get(url)

        # Wait for the page to load
        time.sleep(random.uniform(5, 10))

        # Locate all divs containing links to PDFs
        divs_with_links = driver.find_elements(By.CSS_SELECTOR, 'div.rst-postinfo.bg-color-gray.ext-dk-yl.ext-dk.tx-center')
        print(f"Found {len(divs_with_links)} potential divs on page {page_num}")

        for div in divs_with_links:
            try:
                # Locate the link inside the div
                link = div.find_element(By.TAG_NAME, 'a')
                file_url = link.get_attribute('href')

                # Skip if no URL is found
                if not file_url:
                    continue

                # Check if the link leads to a PDF
                response = requests.head(file_url, allow_redirects=True)
                content_type = response.headers.get('Content-Type', '')

                if 'application/pdf' in content_type:
                    # Generate the PDF file name
                    pdf_name = os.path.join(download_dir, os.path.basename(file_url))
                    # Ensure the filename ends with '.pdf'
                    if not pdf_name.endswith('.pdf'):
                        pdf_name += '.pdf'

                    # Download the PDF
                    response = requests.get(file_url, stream=True)
                    response.raise_for_status()

                    with open(pdf_name, 'wb') as pdf_file:
                        pdf_file.write(response.content)

                    print(f"Downloaded: {pdf_name}")
                    time.sleep(random.uniform(3, 7))
                else:
                    print(f"Skipped non-PDF link: {file_url} (Content-Type: {content_type})")

            except Exception as e:
                print(f"Error processing div: {e}")

    except Exception as e:
        print(f"An error occurred while processing page {page_num}: {e}")

# Close the Selenium WebDriver
driver.quit()

print("Script completed successfully!")
```

```python
import os
import json
from llama_parse import LlamaParse
import nest_asyncio
from typing import Dict, Any

# Apply nest_asyncio to allow async execution in Jupyter or certain environments
nest_asyncio.apply()

# Define a helper function to load or parse PDF data
def load_or_parse_data(pdf_file: str, parsing_instructions: str, llamaparse_api_key: str, language: str) -> Dict[str, Any]:
    parser = LlamaParse(
        api_key=llamaparse_api_key,
        result_type="markdown",  # We want JSON as the result type
        parsing_instruction=parsing_instructions,
        max_timeout=5000,
        language=language,
    )
    parsed_data = parser.load_data(pdf_file)
    return parsed_data

# Function to ensure all data can be serialized to JSON
def convert_to_serializable(parsed_data: Any) -> Any:
    """
    Convert non-serializable objects (like `Document`) to serializable types.
    """
    if isinstance(parsed_data, dict):
        return {key: convert_to_serializable(value) for key, value in parsed_data.items()}
    elif isinstance(parsed_data, list):
        return [convert_to_serializable(item) for item in parsed_data]
    elif hasattr(parsed_data, '__dict__'):  # Handle custom objects with __dict__ attribute
        return parsed_data.__dict__
    else:
        return parsed_data  # Base case: return data as is if it's already serializable

# Define a function to process all PDFs in a folder and save each to a separate JSON file
def convert_pdfs_to_json(pdf_folder: str, json_folder: str, parsing_instructions: str, llamaparse_api_key: str, language: str):
    # Ensure the JSON folder exists
    if not os.path.exists(json_folder):
        os.makedirs(json_folder)

    # Iterate through all files in the pdf_folder
    for filename in os.listdir(pdf_folder):
        pdf_file_path = os.path.join(pdf_folder, filename)

        # Check if the file is a PDF
        if pdf_file_path.lower().endswith('.pdf'):
            print(f"Processing {pdf_file_path}...")

            # Parse the PDF
            parsed_data = load_or_parse_data(pdf_file_path, parsing_instructions, llamaparse_api_key, language)

            # Convert parsed data to a serializable format
            serializable_data = convert_to_serializable(parsed_data)

            # Define the JSON file path
            json_file_path = os.path.join(json_folder, f"{os.path.splitext(filename)[0]}.json")

            # Save parsed data as JSON
            with open(json_file_path, 'w', encoding='utf-8') as json_file:
                json.dump(serializable_data, json_file, ensure_ascii=False, indent=4)

            print(f"Saved parsed data for {filename} to {json_file_path}")

```

```python
import re
import pandas as pd
import os
import json

# Folder path where the JSON files are located
json_folder_path = '/DSI444/data_json'

# List to store all extracted records
data_list = []

# Iterate over all files in the specified folder
for filename in os.listdir(json_folder_path):
    file_path = os.path.join(json_folder_path, filename)

    # Only process .json files
    if filename.endswith('.json'):
        with open(file_path, 'r', encoding='utf-8') as file:
            json_data = file.read()

        # Split the data into individual job listings
        job_listings = re.split(r'### Job Listing', json_data)

        # Loop through each job listing and extract data
        for listing in job_listings:
            if listing.strip():  # Skip empty listings
                occupation = re.search(r'\*\*Occupation\*\*: (.+)', listing)
                description = re.search(r'\*\*Description\*\*: (.+)', listing)
                skill = re.search(r'\*\*Skill\*\*: (.+)', listing)
                income = re.search(r'\*\*Income\*\*: (.+)', listing)
                province = re.search(r'\*\*Province\*\*: (.+)', listing)

                # Append the extracted data to the list if all fields are found
                if occupation and description and skill and income and province:
                    # Clean each field to remove everything after the first newline '\n'
                    occupation_cleaned = occupation.group(1).split("\\n")[0]
                    description_cleaned = description.group(1).split("\\n")[0]
                    skill_cleaned = skill.group(1).split("\\n")[0]
                    income_cleaned = income.group(1).split("\\n")[0]
                    province_cleaned = province.group(1).split("\\n")[0]

                    # Add cleaned data to the list
                    data_list.append({
                        'Occupation': occupation_cleaned,
                        'Description': description_cleaned,
                        'Skill': skill_cleaned,
                        'Income': income_cleaned,
                        'Province': province_cleaned
                    })

# Convert list of dictionaries to DataFrame
df = pd.DataFrame(data_list)

# Optional: Save DataFrame to a CSV file
# df.to_csv('/path/to/save/output.csv', index=False)
#df = df.applymap(lambda x: x.split('\n')[0] if isinstance(x, str) else x)

# Print or use the DataFrame
print(df)
```

```python
driver = GraphDatabase.driver(uri, auth=(username, password))

# Function to insert triples
def create_triples(tx, triples):
    for subject, predicate, obj in triples:
        # Differentiating node types based on the predicate
        if predicate == "liveIn":
            node_label = "Province"
        elif predicate == "hasSkill":
            node_label = "Skill"
        elif predicate == "hasIncome":
            node_label = "Income"
        elif predicate == "hasDescription":
            node_label = "Description"
        else:
            node_label = "Attribute"

        query = (
            f"MERGE (a:Occupation {{name: $subject}}) "
            f"MERGE (b:{node_label} {{value: $object}}) "
            f"MERGE (a)-[:{predicate}]->(b)"
        )
        tx.run(query, subject=subject, object=obj)

# Insert triples into Neo4j
with driver.session() as session:
    session.write_transaction(create_triples, triples)

driver.close()
```
