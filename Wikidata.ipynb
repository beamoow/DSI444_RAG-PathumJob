## Wikidata (Thai)

```python
# Define the SPARQL query
query = """
SELECT ?occupation ?occupationLabel ?description WHERE {
  ?occupation wdt:P31 wd:Q28640;
             rdfs:label ?occupationLabel;
             schema:description ?description.
  FILTER (LANG(?occupationLabel) = "en")
  FILTER (LANG(?description) = "en")
}
LIMIT 10000
"""

# Define the URL for the Wikidata SPARQL endpoint
url = 'https://query.wikidata.org/sparql'

# Send the request to the SPARQL endpoint
response = requests.get(url, params={'query': query, 'format': 'json'})

# Check the response status and process the result into a DataFrame
if response.status_code == 200:
    data = response.json()

    # Extract the relevant information from the API response
    occupations = []
    descriptions = []

    for item in data['results']['bindings']:
        occupation_label = item['occupationLabel']['value']
        description = item['description']['value']

        occupations.append(occupation_label)
        descriptions.append(description)

    time.sleep(2)

    # Create a Pandas DataFrame
    df_occupation = pd.DataFrame({
        'Label': occupations,
        'Description': descriptions
    })

    # Display the DataFrame
    print(df_occupation)
else:
    print(f"Error: {response.status_code}")
```

```python
import pandas as pd
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer
import stanza
import os
from neo4j import GraphDatabase
```

```python
# Load the M2M100 translation model and tokenizer
tokenizer = M2M100Tokenizer.from_pretrained("facebook/m2m100_418M")
model = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M")
```

```python
# Function to translate text from English to Thai
def translate_to_thai(text):
    tokenizer.src_lang = "en"
    encoded = tokenizer(text, return_tensors="pt")
    generated_tokens = model.generate(
        **encoded, forced_bos_token_id=tokenizer.get_lang_id("th")
    )
    return tokenizer.decode(generated_tokens[0], skip_special_tokens=True)
```

```python
# Create triples
triples = []
for index, row in df_combined.iterrows():
    triples.append((row['Label'], "hasDescription", row['Description']))
    triples.append((row['Label'], "isA", row['Type']))

# Print triples
print(triples)
```

```python
driver = GraphDatabase.driver(uri, auth=(username, password))

# Function to insert triples
def create_triples(tx, triples):
    for subject, predicate, obj in triples:

        if predicate == "hasDescription":
          node_label = "Description"
        else:
          node_label = "Type"

        query = (
            f"MERGE (a:Label {{name: $subject}}) "
            f"MERGE (b:{node_label} {{value: $object}}) "
            f"MERGE (a)-[:{predicate}]->(b)"
        )
        tx.run(query, subject=subject, object=obj)

# Insert triples
with driver.session() as session:
    session.write_transaction(create_triples, triples)

driver.close()
```
